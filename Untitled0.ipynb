## Extracting the path

```python
import os
from pathlib import Path

BASE = Path("/content/Hackthon2_scripts/datasets")
print(f"Base datasets folder: {BASE}\n")

def find_candidate_dirs(base):
    cand = []
    for p in base.rglob('*'):
        if p.is_dir():
            name = p.name.lower()
            if any(k in name for k in ['train','val','test']):
                cand.append(p)
    return sorted(set(cand))

def summarize_folder(p):
    imgs = list(p.rglob('*.jpg')) + list(p.rglob('*.png')) + list(p.rglob('*.jpeg'))
    labels = list(p.rglob('*.txt'))
    # filter labels that are next to images: same relative folder or "labels" sibling
    return len(imgs), len(labels), imgs[:3], labels[:3]

candidates = find_candidate_dirs(BASE)
if not candidates:
    print("No train/val/test folders found under", BASE)
else:
    print("Found candidate folders (sample):\n")
    for d in candidates:
        print(" -", d)

print("\nDetailed summary (attempting to map train/val/test):\n")
# heuristics to pick best train/val/test
def pick_dir(kind):
    # prefer exact match, then containing match, then last fallback
    exact = [d for d in candidates if d.name.lower() == kind]
    if exact: return exact[0]
    contain = [d for d in candidates if kind in d.name.lower()]
    if contain: return contain[0]
    return None

train_dir = pick_dir('train') or pick_dir('train_1') or pick_dir('train1')
val_dir   = pick_dir('val')   or pick_dir('val_1')   or pick_dir('val1')
test_dir  = pick_dir('test')  or pick_dir('test3')   or pick_dir('test_3')

for label, d in (('train', train_dir), ('val', val_dir), ('test', test_dir)):
    if d is None:
        print(f"{label.upper():5}: NOT FOUND")
    else:
        imgs, labs, img_sample, lab_sample = summarize_folder(d)
        print(f"{label.upper():5}: {d}")
        print(f"    images: {imgs}    labels: {labs}")
        if img_sample:
            print("    sample images:", [str(x.name) for x in img_sample])
        if lab_sample:
            print("    sample labels:", [str(x.name) for x in lab_sample])

        same_child = next((c for c in d.iterdir() if c.is_dir() and c.name.lower() in d.name.lower()), None)
        if same_child:
            print("    NOTE: Possible double-nested folder found ->", same_child)
        print()

print("Suggested yolo_params paths to use in yolo_params.yaml (copy-paste):\n")
if train_dir:
    print("train:", train_dir)
if val_dir:
    print("val:  ", val_dir)
if test_dir:
    print("test: ", test_dir)

print("\nIf any of the suggestions are wrong, run: !ls -R /content/Hackthon2_scripts/datasets to inspect manually.")
```

## Checking Dataset Structure and Counts

```python
from pathlib import Path

BASE = Path("/content/Hackthon2_scripts/datasets")  # change only if your datasets are elsewhere

def find_image_label_dirs(base):
    results = {}
    for p in base.rglob('*'):
        if p.is_dir() and p.name.lower() in ('images', 'image', 'imgs'):
            labels_dir = p.parent / 'labels'
            key = p.parent.name  # folder containing 'images' (e.g., train_1, val1, test3)
            results[key] = {'images': p, 'labels': labels_dir if labels_dir.exists() else None}
    return results

def count_files(p: Path, exts):
    if p is None or not p.exists():
        return 0
    return sum(1 for f in p.rglob('*') if f.suffix.lower() in exts)

dirs = find_image_label_dirs(BASE)
if not dirs:
    print("No images folder found under", BASE)
else:
    total_imgs = 0
    total_labels = 0
    print(f"{'Split':<15}{'Images':>8}{'Labels':>10}{'Mismatch':>12}{'Example image':>20}")
    for split, paths in dirs.items():
        imgs = count_files(paths['images'], {'.jpg', '.jpeg', '.png'})
        labs = count_files(paths['labels'], {'.txt'}) if paths['labels'] is not None else 0
        total_imgs += imgs
        total_labels += labs
        mismatch = imgs - labs
        example_img = next(paths['images'].rglob('*'), None)
        example_img = example_img.name if example_img is not None else 'N/A'
        print(f"{split:<15}{imgs:8d}{labs:10d}{mismatch:12d}{example_img:20s}")
    print()
    print(f"{'TOTAL':<15}{total_imgs:8d}{total_labels:10d}{(total_imgs-total_labels):12d}")
    if total_imgs != total_labels:
        print("\nNote: image-count and label-count differ. Check label filenames / folders for missing labels.")
```

```python
import os
from IPython.display import Image, display
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Path to your latest run
latest_run = "/content/Hackthon2_scripts/Hackathon2_scripts/runs/space_station_train"

# Display existing images (results, confusion matrix)
for img_name in ["results.png", "confusion_matrix.png"]:
    img_path = os.path.join(latest_run, img_name)
    if os.path.exists(img_path):
        display(Image(filename=img_path))
    else:
        print(f"{img_name} not found in {latest_run}")

# Load trained model for evaluation
model_path = os.path.join(latest_run, "weights", "best.pt")
if os.path.exists(model_path):
    model = YOLO(model_path)
    results = model.val(data="/content/Hackthon2_scripts/Hackathon2_scripts/yolo_params.yaml", split='test')

    # Generate PR curve for class 0 (can loop for all classes if needed)
    if hasattr(results, "pr_curve") and results.pr_curve is not None:
        for class_idx, pr_data in enumerate(results.pr_curve):
            plt.figure(figsize=(6,5))
            plt.plot(pr_data[:,0], pr_data[:,1])
            plt.xlabel("Recall")
            plt.ylabel("Precision")
            plt.title(f"PR Curve - Class {class_idx}")
            plt.grid(True)
            plt.show()
    else:
        print("PR curve data not available in results.")
else:
    print(f"Model weights not found: {model_path}")
```
